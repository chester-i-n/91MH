```python
import torch
import pandas as pd
import numpy as np
import os
import xlrd
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams.update({'font.size': 16})

from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import  mean_squared_error
from sklearn.metrics import r2_score
```

```python
R = 0.01365
R_ = 0.012575
R_C = 0.001
M = 0.000719
L = 0.05037

RPM = np.array([12, 17, 22, 30, 40, 50, 60, 70, 80, 90, 100, 105, 120, 135, 140, 150, 160, 180, 200])
W = np.array([i * 2 * np.pi / 60 for i in RPM])
KSI = np.array([R_ ** 2 * i / (R ** 2 - R_ ** 2) for i in W])


CUT_BEGIN = 10
CUT_END = 6
WORK_DIR = 'work'

path = 'data'
```

```python
def cut_df(df):
    start = 0
    tmp_df = []
    for i in range(1, len(df)):
        if df.loc[i, ' Speed'] != df.loc[i - 1, ' Speed']:
            if df.loc[start, ' Speed']:
                tmp_df.append(df.iloc[start + CUT_BEGIN:i - CUT_END])
            start = i
    if df.loc[start, ' Speed'] > 0:
        tmp_df.append(df.iloc[start + CUT_BEGIN:i - CUT_END])
    return pd.concat(tmp_df)
```

```python
tmp_df = []
for name in os.listdir(path):
    if "XLS" not in name:
        continue
    wb = xlrd.open_workbook(f"{path}/{name}", encoding_override='iso-8859-1')
    df = pd.read_excel(wb)
    df.columns = df.iloc[3].to_list()
    df = df.iloc[5:-1, 1:-1].reset_index(drop=True).astype(float)
    df['Torque'] /= 100
    df.insert(1, "Angular velocity", (df.iloc[:, 1] * 2 * np.pi) / 60)
    df.loc[:, "Shear Stress"] = M * df.loc[:, "Torque"] / (2 * np.pi * R_ ** 2 * L)
    df.loc[:, "Shear Rate"] = R_ ** 2 * df.loc[:, "Angular velocity"] / (R ** 2 - R_ ** 2)
    df.loc[:, "Viscosity"] = df.loc[:, "Shear Stress"] / (2 * df.loc[:, "Shear Rate"])
    plt.figure(figsize=(16, 10))
    plt.plot(df['Torque'].to_numpy() * 350)
    plt.plot(df[' Speed'].to_numpy())
    plt.title('Before cut')
    plt.legend(['Torque', 'Speed'])
    df = cut_df(df)
    plt.figure(figsize=(16, 10))
    plt.plot(df['Torque'].to_numpy() * 350)
    plt.plot(df[' Speed'].to_numpy())
    plt.title('After cut')
    plt.legend(['Torque', 'Speed'])
    plt.show()
    display(df)
    tmp_df.append(df)
    df.to_csv(f"{path}/{name[:-4]}.csv")
```

```python
df_all = pd.DataFrame(columns=df.columns)
for i in range(len(tmp_df)):
    df_all.loc[i] = [list(tmp_df[i][j]) if " C" not in j else tmp_df[i][j][0] for j in tmp_df[i].columns]

df_all.to_csv(f"{path}/all_data.csv")
display(df_all)

df_all_reg = pd.concat(tmp_df)
df_all_reg.to_csv(f"{path}/all_data_reg.csv")
display(df_all_reg)
```

```python
class RegDataset(torch.utils.data.Dataset):
    
    def __init__(self, x, y=None):
        self.x = x
        self.y = y.unsqueeze(1) if y is not None else self.x

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return (self.x[idx], self.y[idx])


class Osvald(torch.nn.Module):

    def __init__(self):
        super(Osvald, self).__init__()
        self.alpha_list = torch.nn.ParameterList([])            
        self.q = torch.nn.parameter.Parameter(torch.randn(1, requires_grad=True))
        self.alpha_list.append(self.q)
        self.z = torch.nn.parameter.Parameter(torch.tensor(0.9, requires_grad=True))
        self.alpha_list.append(self.z)

    def forward(self, x):

        return self.q * x ** (self.z - 1)

    def predict(self, loader):
        self.eval()
        pred = []
        for x_batch, y_batch in loader:
                
            pred.append(self.forward(x_batch))
        
        pred = torch.cat(pred, axis=0)
        return pred.detach().cpu().numpy()
    
    def evaluate(self, loader, criterion):
        self.eval()
        pred = []
        labels = []
        for x_batch, y_batch in loader:
                
            pred.append(self.forward(x_batch))
            labels.append(y_batch)
        
        
        return (criterion(torch.cat(pred, axis=0), torch.cat(labels, axis=0)),
                    r2_score(torch.cat(labels, axis=0), torch.cat(pred, axis=0).detach()))
    
    def fit(self, loader, optimizer, criterion):
        self.train()
        history = []

        for i in range(1000):
            run_loss = 0
            iters = 0

            for x_batch, y_batch in loader:

                optimizer.zero_grad()

                out = self.forward(x_batch)

                loss = criterion(out, y_batch)
                loss.backward()

                optimizer.step()

                iters += 1

                run_loss += loss.item()

            history.append(run_loss / iters)
    
        return history

    
os.makedirs(WORK_DIR + '/models', exist_ok=True)
    
criterion = torch.nn.MSELoss()
    
markers = ['o', 'v', '^', '<', '>']

x = np.expand_dims(df_all_reg['Shear Rate'], axis=1)

y = df_all_reg['Viscosity']
x_vis = np.expand_dims(np.unique(df_all_reg['Shear Rate']), axis=1)
    
X = torch.tensor(x, dtype=torch.float32)
Y = torch.tensor(y.to_numpy(), dtype=torch.float32)

X_vis = torch.tensor(x_vis, dtype=torch.float32)

dataset = RegDataset(X, Y)

dataset_vis = RegDataset(X_vis)

loader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True)

loader_vis = torch.utils.data.DataLoader(dataset_vis, batch_size=32)

loss = -1
r2 = -1

for lr in [10, 0.1, 0.001]:
    
    model = Osvald()

    optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    
    model.fit(loader, optimizer, criterion)
    
    cur_loss, cur_r2 = model.evaluate(loader, criterion)
    
    print(f"lr: {lr}, loss: {cur_loss:.5f}, r2 score: {cur_r2:.5f}")
    
    if loss == -1 or cur_loss < loss:
        if torch.isnan(cur_loss):
            loss = -1
            r2 = -1
        else:
            loss = cur_loss
            r2 = cur_r2
        torch.save(model.state_dict(), f"{WORK_DIR}/models/model.pth")
        best_lr = lr

reg_model = Osvald()
reg_model.load_state_dict(torch.load(f"{WORK_DIR}/models/model.pth"))

y_vis = reg_model.predict(loader_vis)

plt.figure(figsize=(16, 8))

legend_list = []

visc = df_all.apply(lambda x: pd.Series(x['Viscosity']), axis=1).stack().reset_index(level=1, drop=True)

ksi = df_all.apply(lambda x: pd.Series(x['Shear Rate']), axis=1).stack().reset_index(level=1, drop=True)

df2 = pd.DataFrame({'Viscosity': visc, 'Shear Rate': ksi})

plt.plot(x_vis, y_vis)
for i in range(df_all.shape[0]):
    plt.scatter(df_all.loc[i, "Shear Rate"], df_all.loc[i, "Viscosity"], marker=markers[i], s=100, alpha=.5)

plt.xlabel(r'$\xi$, Shear rate (1/s)')
plt.ylabel(r'$\mu$, Dynamic viscosity ($Pas$)')
plt.legend([r'${:.3f}\xi^{}$'.format(reg_model.q.item(), '{' + f'{reg_model.z.item():.3f}'+ '-1}')]
           + [f'{j} parallel test' for j in range(1, df_all.shape[0] + 1)])
plt.title(r'$R^2$: ' + f'{r2:.3f}')
plt.show()
```
